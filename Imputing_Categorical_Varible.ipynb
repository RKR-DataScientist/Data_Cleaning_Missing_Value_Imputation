# Imputing Categorical Varible Using MODE


```
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
# Load Dataset
df = pd.read_csv(r'/Train.csv')
df.head()
```

```
# To see all the columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
df.head()
```

```
df.shape
```

```
# To select all the categorial data from dataset
cat_vars = df.select_dtypes(include='object')
cat_vars.head()
```

```
# To check the percentage of missing data in categorical columns
miss_percent = cat_vars.isnull().mean()*100
miss_percent
```

```
# Let's select those varible whose percentage more than 20%, will drop that
drop_vars = ['Alley','FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
cat_vars.drop(columns=drop_vars, axis = 1, inplace= True)
cat_vars.shape
```

```
# Now we will check all the missing values more than 0
isnull_per = cat_vars.isnull().mean()*100
miss_vars = isnull_per[isnull_per > 0]
miss_vars
```

```
# values comes in series, but we want only object values
miss_vars = isnull_per[isnull_per > 0].keys()
miss_vars
```

```
# If we want to check the mode values of any columns
cat_vars['MasVnrType'].mode()
```

```
# None is more in columns
cat_vars['MasVnrType'].value_counts()
```

## Imputing with Mode 

```
# Let's try with one column, will take 0 index, beacuse none is in 0 index
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].value_counts()[0])
```

```
# To fill all the categorical varible
for var in miss_vars:
  cat_vars[var].fillna(cat_vars[var].mode()[0], inplace = True)
  print(var,'=', cat_vars[var].mode()[0])
```

```
# To check the categorical varible has been filled or not
cat_vars.isnull().sum()
```

## Histogram Plot 
TO check difference between original and imputed dataset.


```
plt.figure(figsize=(16,9))
for i, var in enumerate(miss_vars):
  plt.subplot(4, 3, i+1)
  plt.hist(cat_vars[var], label='Imputed')
  plt.hist(df[var].dropna(), label = 'Original')
  plt.legend()
```

You can see the imputed and orginal data has not big differences. Both distribution is same on scale

```
# Now we want to update our dataset 
df.update(cat_vars)
```

```
# Let's drop that features whose more than 20% missing from dataset
df.drop(columns=drop_vars, inplace=True)
```

```
# Now we want to check wheather data imputation has properly imputed or not
df.select_dtypes(include='object').isnull().sum()
```
